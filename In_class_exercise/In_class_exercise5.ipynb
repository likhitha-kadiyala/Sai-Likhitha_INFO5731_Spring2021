```
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('-headless')
chrome_options.add_argument('-no-sandbox')
chrome_options.add_argument('-disable-dev-shm-usage')
```

```
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.by import By
import pandas as pd
import time

web_driver = webdriver.Chrome('chromedriver',options=chrome_options)
url = 'https://www.imdb.com/title/tt7286456/reviews?ref_=tt_urv'
title = []
reviews = []
web_driver.get(url)
for value in range(20):
  web_driver.find_element_by_class_name("ipl-load-more__button").click()
  time.sleep(5)
  title_list = web_driver.find_elements(By.CLASS_NAME, "title")
  reviews_list = web_driver.find_elements(By.CLASS_NAME, "text")
for tit, rev in zip(title_list, reviews_list):
      title.append((tit.text).replace('\n',''))
      reviews.append(rev.text)
imdb_df = pd.DataFrame(list(zip(title, reviews)), columns =['Title', 'Reviews'])
print("Length of data frame is {0}".format(len(imdb_df)))
imdb_df
```

# **Preprocessing the data**

**Lower case**

```
imdb_df['lower case'] = imdb_df['Reviews'].apply(lambda y: " ".join(y.lower() for y in y.split()))
```

**Punctuation removal**

```
imdb_df['text_preprocessing'] = imdb_df['lower case'].str.replace('[^\w\s]','')
imdb_df
```

**Digits Removal**

```
imdb_df['text_preprocessing'] = imdb_df['text_preprocessing'].apply(lambda y: ''.join([i for i in y if not i.isdigit()]))
```

**Special characters removal**

```
import re
imdb_df['text_preprocessing'] = imdb_df['text_preprocessing'].apply(lambda y: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', char) for char in y ))
imdb_df
```

```
import nltk
nltk.download()
```

**REmoving stopwords**

```
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
imdb_df['text_preprocessing'] = imdb_df['text_preprocessing'].apply(lambda y: " ".join(y for y in y.split() if y not in stop_words))
imdb_df
```

**Stemming**

```
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
imdb_df['stemming'] = imdb_df['lower case'].apply(lambda y: " ".join([stemmer.stem(value) for value in y]))
imdb_df
```

**Lemmatization**

```
from textblob import Word
import nltk
nltk.download('wordnet')

imdb_df['lemmatization'] = imdb_df['lower case'].apply(lambda y: " ".join([Word(value).lemmatize() for value in y.split()]))
imdb_df
```

**Parts of speech tagging and features**

```
from nltk.tokenize import word_tokenize
POS = []
for sent in imdb_df['lower case']: 
  word = word_tokenize(sent)
  POS.append(nltk.pos_tag(word))
POS
```

```
Adjective = []
Adverb = []
CordinatingConjunction = []
preposition = []
Interjection = []
Noun = []
Verb = []
PersonalPronoun = []
predeterminer = []
Determiner = []

for value in POS:
  Adjective_Count = 0
  Adverb_Count = 0
  CordinatingConjunction_Count = 0
  preposition_Count = 0
  Interjection_Count = 0
  Noun_Count = 0
  Verb_Count = 0
  PersonalPronoun_Count = 0
  predeterminer_Count = 0
  Determiner_Count = 0
  for word,tag in value:
    if tag == 'JJ':
      Adjective_Count = Adjective_Count + 1
    elif tag == 'RB':
      Adverb_Count = Adverb_Count + 1
    elif tag == 'CC':
      CordinatingConjunction_Count = CordinatingConjunction_Count + 1
    elif tag == 'IN':
      preposition_Count = preposition_Count + 1
    elif tag == 'UH':
      Interjection_Count = Interjection_Count + 1
    elif tag == 'NN':
      Noun_Count = Noun_Count + 1
    elif tag == 'VB':
      Verb_Count = Verb_Count + 1
    elif tag == 'PRP':
      PersonalPronoun_Count = PersonalPronoun_Count + 1
    elif tag == 'PDT':
      predeterminer_Count = predeterminer_Count + 1
    elif tag == 'DT':
      Determiner_Count = Determiner_Count + 1
  Adjective.append(Adjective_Count)
  Adverb.append(Adverb_Count)
  CordinatingConjunction.append(CordinatingConjunction_Count)
  preposition.append(preposition_Count)
  Interjection.append(Interjection_Count)
  Noun.append(Noun_Count)
  Verb.append(Verb_Count)
  PersonalPronoun.append(PersonalPronoun_Count)
  predeterminer.append(predeterminer_Count)
  Determiner.append(Determiner_Count)
```

```
imdb_df['No.of Adjectives'] = Adjective
imdb_df['No.of Adverbs'] = Adverb
imdb_df['No.of Cordinating Conjunctions'] = CordinatingConjunction
imdb_df['No.of prepositions'] = preposition
imdb_df['No.of Interjections'] = Interjection
imdb_df['No.of Nouns'] = Noun
imdb_df['No.of Verbs'] = Verb
imdb_df['No.of Personal Pronouns'] = PersonalPronoun
imdb_df['No.of Predeterminers'] = predeterminer
imdb_df['No.of Determiners'] = Determiner
imdb_df

```

**Linguistic Features**

**Number of right branching nodes**

```
import spacy
Nodes = []
nlp = spacy.load("en_core_web_sm")
for sent in imdb_df['lower case']:
  doc = nlp(sent)
  try:
    Nodes.append(doc[0].n_rights)
  except:
    Nodes.append('No')
imdb_df['Number of Right Branching Nodes'] = Nodes
```

**Length of sentence**

```
imdb_df['Sentence Length'] = imdb_df['Reviews'].apply(lambda y: len(y))
imdb_df
```
